```python
print("========================================================================")
print("               Spam Message Detection System                            ")
print("========================================================================")
message= input("Enter a message to check ")
message=message.lower()
spam_keywords=["win","free","prize","offer","click","urgent","limited","congratulations","lottery","claim"]
spam_message=False
for word in spam_keywords:
    if word in message:
        spam_message=True
        break
        spam_message=False
if spam_message:
    print("This message is identified as spam message")
else:
  print("This message is not identified as spam message")
  print("\n Thank you for using Spam detection message System")
```
